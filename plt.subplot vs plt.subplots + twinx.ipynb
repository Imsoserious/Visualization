# 한 번에 여러 개의 그래프를 출력하는 방법
#### - matplotlib에서는 subplot, subplots이 쓰인다.
    - plt.subplot(nrows, ncols, index) : 그래프 위치를 각각 지정해줘야 한다.
    - plt.subplots(nrows, ncols) : 그래프 위치를 한 번에 지정할 수 있다.

**1) plt.subplot()**

```python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

- 각 그래프마다 plt.subplot() 코드를 작성해서 index를 지정해주어야 한다.

**2) plt.subplots()**

```python
fig, axes = plt.subplots(2, 2) # fig는 전체 그래프, axes는 개별 그래프 객체
```

```python
print(fig)
print('\n')
print(axes, axes.shape)
```

```python
fig, axes = plt.subplots(2, 2) # fig는 전체 그래프, axes는 개별 그래프 객체

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")
fig.suptitle('figure title', fontsize=15) # fig의 제목 설정은 subtitle, axes의 제목 설정은 set_title

plt.tight_layout()
plt.show()
```

- 그래프 제목이나 축제목 등 설정 시에는 'set_'을 붙인다.

**3) twinx()**
- 2개의 그래프가 x축을 공유하게 해준다.

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx() # 
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)
ax0.set_xlabel("공유되는 x축")
plt.show()
```
